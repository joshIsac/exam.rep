```python
from sklearn.preprocessing import OneHotEncoder
cat_columns=['CRIM','ZN','INDUS','CHAS','AGE','LSTAT']
data_encoded=pd.get_dummies(data ,columns=cat_columns,drop_first=True)
print(data_encoded)
```

```python
plt.bar(data['INDUS'] ,data['CRIM'])
plt.xlabel("no of rooms")
plt.ylabel("crime rate")
plt.title("relationships between housing attributes")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sn.scatterplot(x='RM', y='MEDV', data=data)
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median House Price (MEDV)')
plt.show()
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X= data_encoded.drop('MEDV', axis=1)  # Features
y = data_encoded['MEDV']              # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model =LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # Taking the square root of MSE to get RMSE
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R-squared (R2): {r2}")
```
